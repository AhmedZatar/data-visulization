# Visualization with Seaborn Class

## Import seaborn to plot graph

```python
import seaborn as sns
```

## Get Dataset using seaborn

```python
cars_mpg = sns.load_dataset('mpg')
cars_mpg_usa=cars_mpg[cars_mpg['origin']=='usa']
cars_mpg_europe=cars_mpg[cars_mpg['origin']=='europe']
cars_mpg_japan=cars_mpg[cars_mpg['origin']=='japan']
cars_mpg
```

## Give theme for tha graph

```python
sns.set()
```

## MPG vs Horsepower for cars made in USA

```python
sns.relplot(x='horsepower', y='mpg', kind='line',color = 'red', data=cars_mpg_usa)
```

## MPG vs Horsepower for cars made in Japan

```python
sns.relplot(x='horsepower', y='mpg', kind='line',color = 'orange', data=cars_mpg_japan)
```

## MPG vs Horsepower for cars made in Europe

```python
sns.relplot(x='horsepower', y='mpg', kind='line',color = 'blue', data=cars_mpg_europe)
```
