# Visualization with Seaborn Class

## Import seaborn to plot graph

```python
import seaborn as sns
```

## Get Dataset using seaborn

```python

tips = sns.load_dataset('tips')
female_tips=tips[tips['sex']=='Female']
male_tips=tips[tips['sex']=='Male']
tips
```

## Give theme for tha graph

```python
sns.set()
```

## Tip vs Total_bill for female graph

```python

sns.relplot(x='total_bill', y='tip', kind='line',color = 'red', data=female_tips)
```

## Tip vs Total_bill for male graph

```python
sns.relplot(x='total_bill', y='tip', kind='line',color = 'blue', data=male_tips)
```

# Conclusions

### 1- In the female graph as you see at first when bill increase tip increase but when the bill become too high the tip decrease

### 2- In the male graph as you see at first when bill increase tip increase even when the bill become too high the tip increase 

### 3- Male got higher tip than female tip
